```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.e**(-x/4) * np.arctan(x)

# Lite tall for numerisk tilnærming
h = 1e-6

# Tilnærmer f'(x) med definisjonen av den deriverte
def d(x, h):
    return (f(x+h) - f(x)) / h

# Newtons metode: finn x der f'(x) = 0
# d(x) er f'(x) og ((d(x + h,h) - d(x,h)) / h er f''(x) for å finne endringen i d(x)
x = 1.0 # Startgjett
for i in range(100):
    x = x - d(x,h) / ((d(x + h,h) - d(x,h)) / h)   # Flytter x nærmere nullpunktet til f'(x) for hver iterasjon

x_topp = x
f_topp = f(x_topp)

print(x_topp)
print(f_topp)

# Plotter funksjonen
x = np.linspace(-6, 12, 100)
y = f(x)

plt.scatter(x_topp, f_topp, label="topp_punkt", color="red") # Toppunktet til f(x)
plt.plot(x, y, label="f(x)") # Funksjonen: f(x)
plt.xlabel("x") 
plt.ylabel("y")
plt.title("funksjon")
plt.legend()
plt.grid(True)
plt.show()
```
